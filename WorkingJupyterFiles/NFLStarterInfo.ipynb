```python
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
teamabbr = ['arz','atl','bal','buf','car','chi','cin','cle','dal','den','det','gb','hou',
           'ind','jac','kc','lac','lam','mia','min','ne','no','nyg','nyj','oak','phi','pit',
           'sf','sea','tb','ten','was','stl','sd']
years = ['2009','2011','2012','2013','2014','2015','2016','2017','2018','2019']
dfs=[]
for yr in years:
    for team in teamabbr:
        url = f"http://www.jt-sw.com/football/pro/rosters.nsf/Annual/{yr}-{team}"
        try:
            df = pd.read_html(url)
            new_df = df[1].drop(columns="Unnamed: 2")
            new_df["year"] = yr
            new_df["team"] = team
            dfs.append(new_df)
            print(f"success: {team}:{yr}")
            if len(dfs)==1:
                comb_df = new_df
            else:
                comb_df = pd.concat([comb_df,new_df])
                
        except: print(url)
```
