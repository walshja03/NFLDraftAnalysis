```python
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
import statsmodels.api as sm
import sys
```

```python
Rec2019= pd.read_csv("ReceivingStatsData/2019R.csv")

null_columns=Rec2019.columns[Rec2019.isnull().any()]
Rec2019[null_columns].isnull().sum()
```

```python
#Rec2019.loc[[39,50,166,167,176,186,241,,'Pos']='WR'
print(Rec2019[Rec2019['Pos'].isnull()].head(30))
```

```python
#Create list to store players names for cleaning up
split_player = Rec2019["Player"]
#Split out and store the name of the player
players=[i.split("\\")[0] for i in split_player]
#Split out and store first and last names
fname = [i.split(maxsplit = 1)[0] for i in players]
lname = [i.split(maxsplit = 1)[1] for i in players]
#Add first and last name columns
Rec2019["First Name"]=fname
Rec2019["Last Name"]=lname
```

```python
Rec2019[['Rk','First Name','Last Name','Tm','Pos','G','GS','Tgt','Rec','Yds','1D','YBC','YBC/R','YAC','YAC/R','BrkTkl','Rec/Br','Drop','Drop%']]

Rec2019['Pos'].replace('wr','WR',inplace = True)
Rec2019['Pos'].replace('te','TE',inplace = True)
Rec2019['Pos'].replace(['rb','FB/rb','fb/rb','fb'],'RB',inplace = True)
```

```python
Positions = Rec2019['Pos'].unique()
Positions =['WR', 'RB', 'TE']
Positions
```
